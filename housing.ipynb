```python
import pandas as  pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,r2_score
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
df=pd.read_csv("Housing.csv")
df.columns = df.columns.str.strip()

df=pd.get_dummies(df,columns=["furnishingstatus"],drop_first=True)
yes_no_column=["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea"]
for i in yes_no_column:
    df[i]=df[i].map({"yes":1,"no":0})
true_false_columns=["furnishingstatus_semi-furnished","furnishingstatus_unfurnished"]
for i in true_false_columns:
    df[i]=df[i].map({True:1,False:0})
Y=df["price"]
X=df.drop("price",axis=1)
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.45,random_state=42)

model=LinearRegression()
model.fit(train_x,train_y)


p=model.predict(test_x)
#p=pA*test_x["area"]
print("predicted prices are:")
print(p[:5])
print("Actual prices are:")
print(test_y[:5])



mse=mean_squared_error(test_y,p)
rmse=mse**0.5
mae=mean_absolute_error(test_y,p)
r2=r2_score(test_y,p)
k=X.shape[1]
n=len(Y)
adjusted_r2=1 - (((1 - r2) * (n - 1)) / (n - k - 1))

print("MSE is:",mse)
print("RMSE is:",rmse)
print("MAE is:",mae)
print("R square score is:",r2)
print("adjusted R square score is:",adjusted_r2)
print(model.score(train_x,train_y))
print(model.score(test_x,test_y))
print("difference b/w train and test score is ",model.score(train_x,train_y)-model.score(test_x,test_y))


ressidue=test_y-p
plt.scatter(p,ressidue,color="blue")
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("predicted value")
plt.ylabel("actul value")
plt.show()
```
